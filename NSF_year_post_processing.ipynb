```python
import pandas as pd
import numpy as np
```

```python
NSF_year = pd.read_csv('./NSF_year_wordcount.csv')
```

```python
def clean_string(strng):
    from ast import literal_eval
    return literal_eval(strng[8:-1])

clean_string(NSF_year.loc[0, 'word_count'])
```

```python
NSF_year['word_countdict'] = NSF_year.word_count.apply(clean_string)
```

```python
NSF_year.word_countdict[2]
NSF_year2 = NSF_year[['year', 'word_countdict']]
```

```python
NSF_allwords = pd.DataFrame(columns = ["word", "number", "year"])

for y in list(NSF_year2['year']):
    print(y)
    d = dict()
    d['word'] = list(NSF_year2[NSF_year2['year']==y]['word_countdict'][y-2005].keys())
    d['number'] = list(NSF_year2[NSF_year2['year']==y]['word_countdict'][y-2005].values())
    d['year'] = y
    NSF_allwords = pd.concat([NSF_allwords, pd.DataFrame.from_dict(d)], axis=0)
```

```python
NSF_allwords['number'] = NSF_allwords['number'].astype(int)
pvt = NSF_allwords.pivot_table(values='number', index=['word'], columns=['year']).fillna(value=0).reset_index()
```

```python
pvt.head()
```

```python
for y in list(NSF_year2['year']):
    print(y)
    pvt[str(y)+'m'] = pvt[y] / sum(pvt[y])
```

```python
pvt['distance'] = pvt['2007m'] - pvt['2017m']
pvt_sortasc = pvt.sort_values(by='distance')
pvt_sortdesc = pvt.sort_values(by='distance', ascending=False)
```
