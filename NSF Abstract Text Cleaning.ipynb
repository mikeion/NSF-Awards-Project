This script parses the text from the .xml files downloaded from nsf website!

```python
import pandas as pd
import re
import os

# get list of all .xml files in downloads folder 
xmlfiles = [os.path.join(root, name)
            for root, dirs, files in os.walk('./downloads')
            for name in files
            if name.endswith(('.xml'))]
```

Now, create a loop to gather all the information from the .xml files.

```python
# append with columns from xml file 

# create empty list for appending
lst = []

# specify column names
cols = ['award_title', 
        'award_effective_date',
        'award_expiration_date',
        'award_amount',
        'award_instrument',
        'organization_code',
        'directorate_longname',
        'division_longname',
        'org_program_officer',
        'abstract_text',
        'minamd_letter_date',
        'maxamd_letter_date',
        'award_id',
        'first_name',
        'last_name',
        'email',
        'start_date',
        'role_code',
        'institution',
        'city_name',
        'zipcode',
        'phonenumber',
        'street_address',
        'country',
        'state',
        'state_code',
        'program_code',
        'program_name']

# for loop
for i in range(len(xmlfiles)):
    
    # open current file in year list
    with open(xmlfiles[i], 'r', encoding = 'utf8', errors = 'ignore') as f:
        content = f.read()

    # create blank temp dataframe
    temp = pd.DataFrame()

    # extract each variable from xml and put into blank dataframe
    # award_title
    award_title = ''.join(re.findall('[<]AwardTitle[>].*[<]', content))
    award_title = ''.join(re.sub('[<].*[>]', '', award_title))
    award_title = ''.join(re.sub('[<]', '', award_title))

    # award_effective_date
    award_effective_date = ''.join(re.findall('[<]AwardEffectiveDate[>].*[<]', content))
    award_effective_date = ''.join(re.sub('[<].*[>]', '', award_effective_date))
    award_effective_date = ''.join(re.sub('[<]', '', award_effective_date))

    # award_expiration_date
    award_expiration_date = ''.join(re.findall('[<]AwardExpirationDate[>].*[<]', content))
    award_expiration_date = ''.join(re.sub('[<].*[>]', '', award_expiration_date))
    award_expiration_date = ''.join(re.sub('[<]', '', award_expiration_date))

    # award_amount
    award_amount = ''.join(re.findall('[<]AwardAmount[>].*[<]', content))
    award_amount = ''.join(re.sub('[<].*[>]', '', award_amount))
    award_amount = ''.join(re.sub('[<]', '', award_amount))

    # award_instrument
    award_instrument = ''.join(re.findall('[<]AwardInstrument[>]\n[ ]*[<]Value[>].*[<]', content))
    award_instrument = ''.join(re.sub('[<].*[>]', '', award_instrument))
    award_instrument = ''.join(re.sub('[<]', '', award_instrument))
    award_instrument = ''.join(re.sub('\n[ ]*', '', award_instrument))

    # organization_code
    organization_code = ''.join(re.findall('[<]Organization[>]\n[ ]*[<]Code[>].*[<]', content))
    organization_code = ''.join(re.sub('[<].*[>]', '', organization_code))
    organization_code = ''.join(re.sub('[<]', '', organization_code))
    organization_code = ''.join(re.sub('\n[ ]*', '', organization_code))

    # directorate_longname
    directorate_longname = ''.join(re.findall('[<]Directorate[>]\n[ ]*[<]LongName[>].*[<]', content))
    directorate_longname = ''.join(re.sub('[<].*[>]', '', directorate_longname))
    directorate_longname = ''.join(re.sub('[<]', '', directorate_longname))
    directorate_longname = ''.join(re.sub('\n[ ]*', '', directorate_longname))

    # division_longname
    division_longname = ''.join(re.findall('[<]Division[>]\n[ ]*[<]LongName[>].*[<]', content))
    division_longname = ''.join(re.sub('[<].*[>]', '', division_longname))
    division_longname = ''.join(re.sub('[<]', '', division_longname))
    division_longname = ''.join(re.sub('\n[ ]*', '', division_longname))

    # org_program_officer
    org_program_officer = ''.join(re.findall('[<]\/Organization[>]\n[ ]*[<]ProgramOfficer[>]\n[<]SignBlockName[>].*[<]', content))
    org_program_officer = ''.join(re.sub('[<].*[>]', '', org_program_officer))
    org_program_officer = ''.join(re.sub('[<]', '', org_program_officer))
    org_program_officer = ''.join(re.sub('\n[ ]*', '', org_program_officer))

    # abstract_text
    abstract_text = ''.join(re.findall('[<]AbstractNarration[>].*[<]', content))
    abstract_text = ''.join(re.sub('[<].*[>]', '', abstract_text))
    abstract_text = ''.join(re.sub('[<]', '', abstract_text))

    # minamd_letter_date
    minamd_letter_date = ''.join(re.findall('[<]MinAmdLetterDate[>].*[<]', content))
    minamd_letter_date = ''.join(re.sub('[<].*[>]', '', minamd_letter_date))
    minamd_letter_date = ''.join(re.sub('[<]', '', minamd_letter_date))

    # maxamd_letter_date
    maxamd_letter_date = ''.join(re.findall('[<]MaxAmdLetterDate[>].*[<]', content))
    maxamd_letter_date = ''.join(re.sub('[<].*[>]', '', maxamd_letter_date))
    maxamd_letter_date = ''.join(re.sub('[<]', '', maxamd_letter_date))

    # award_id
    award_id = ''.join(re.findall('[<]AwardID[>].*[<]', content))
    award_id = ''.join(re.sub('[<].*[>]', '', award_id))
    award_id = ''.join(re.sub('[<]', '', award_id))

    # first_name
    first_name = ''.join(re.findall('[<]FirstName[>].*[<]', content))
    first_name = ''.join(re.sub('[<].*[>]', '', first_name))
    first_name = ''.join(re.sub('[<]', '', first_name))

    # last_name
    last_name = ''.join(re.findall('[<]LastName[>].*[<]', content))
    last_name = ''.join(re.sub('[<].*[>]', '', last_name))
    last_name = ''.join(re.sub('[<]', '', last_name))

    # email
    email = ''.join(re.findall('[<]EmailAddress[>].*[<]', content))
    email = ''.join(re.sub('[<].*[>]', '', email))
    email = ''.join(re.sub('[<]', '', email))

    # start_date
    start_date = ''.join(re.findall('[<]StartDate[>].*[<]', content))
    start_date = ''.join(re.sub('[<].*[>]', '', start_date))
    start_date = ''.join(re.sub('[<]', '', start_date))

    # role_code
    role_code = ''.join(re.findall('[<]RoleCode[>].*[<]', content))
    role_code = ''.join(re.sub('[<].*[>]', '', role_code))
    role_code = ''.join(re.sub('[<]', '', role_code))

    # institution
    institution = ''.join(re.findall('[<]Institution[>]\n[ ]*[<]Name[>].*[<]', content))
    institution = ''.join(re.sub('[<].*[>]', '', institution))
    institution = ''.join(re.sub('[<]', '', institution))
    institution = ''.join(re.sub('\n[ ]*', '', institution))

    # city_name
    city_name = ''.join(re.findall('[<]CityName[>].*[<]', content))
    city_name = ''.join(re.sub('[<].*[>]', '', city_name))
    city_name = ''.join(re.sub('[<]', '', city_name))

    # zipcode
    zipcode = ''.join(re.findall('[<]ZipCode[>].*[<]', content))
    zipcode = ''.join(re.sub('[<].*[>]', '', zipcode))
    zipcode = ''.join(re.sub('[<]', '', zipcode))

    # phonenumber 
    phonenumber = ''.join(re.findall('[<]PhoneNumber[>].*[<]', content))
    phonenumber = ''.join(re.sub('[<].*[>]', '', phonenumber))
    phonenumber = ''.join(re.sub('[<]', '', phonenumber))

    # street_address 
    street_address = ''.join(re.findall('[<]StreetAddress[>].*[<]', content))
    street_address = ''.join(re.sub('[<].*[>]', '', street_address))
    street_address = ''.join(re.sub('[<]', '', street_address))

    # country 
    country = ''.join(re.findall('[<]CountryName[>].*[<]', content))
    country = ''.join(re.sub('[<].*[>]', '', country))
    country = ''.join(re.sub('[<]', '', country))

    # state 
    state = ''.join(re.findall('[<]StateName[>].*[<]', content))
    state = ''.join(re.sub('[<].*[>]', '', state))
    state = ''.join(re.sub('[<]', '', state))

    # state_code
    state_code = ''.join(re.findall('[<]StateCode[>].*[<]', content))
    state_code = ''.join(re.sub('[<].*[>]', '', state_code))
    state_code = ''.join(re.sub('[<]', '', state_code))

    # program_code
    program_code = ''.join(re.findall('[<]ProgramElement[>]\n[ ]*[<]Code[>].*[<]', content))
    program_code = ''.join(re.sub('[<].*[>]', '', program_code))
    program_code = ''.join(re.sub('[<]', '', program_code))
    program_code = ''.join(re.sub('\n[ ]*', '', program_code))

    # program_name
    program_name = ''.join(re.findall('[<]Text[>].*[<]', content))
    program_name = ''.join(re.sub('[<].*[>]', '', program_name))
    program_name = ''.join(re.sub('[<]', '', program_name))
    program_name = ''.join(re.sub('\n[ ]*', '', program_name))
  
    # create for loop to append to list
    lst.append([
        award_title, 
        award_effective_date,
        award_expiration_date,
        award_amount,
        award_instrument,
        organization_code,
        directorate_longname,
        division_longname,
        org_program_officer,
        abstract_text,
        minamd_letter_date,
        maxamd_letter_date,
        award_id,
        first_name,
        last_name,
        email,
        start_date,
        role_code,
        institution,
        city_name,
        zipcode,
        phonenumber,
        street_address,
        country,
        state,
        state_code,
        program_code,
        program_name])
    
# create final dataframe from list
data = pd.DataFrame(lst, columns = cols)

# check missing data
data.isnull().sum()
```

Save dataframe as a .csv file for later use.

```python
# save dataframe to .csv
data.to_csv('NSFdata.csv', sep=',')

```
